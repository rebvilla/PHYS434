```python
#2. Using Rayleigh distribution. The parameter for x axis is adjusted from the parameter for normal distribution in HW1. 
d = stats.rayleigh.rvs(loc = 0, scale = 0.01, size = 100000)
```

```python
fig, ax = plt.subplots(1, 1)
ax.hist(d,50, density=True)
plt.tick_params(labelsize = 24)
plt.xlim([0,0.1])
x = np.linspace(0,0.1,1000)
ax.plot(x,stats.rayleigh.pdf(x,loc = 0, scale = 0.01),linewidth = 8,alpha = 0.7)
plt.show()
```
